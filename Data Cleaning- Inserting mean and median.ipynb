```python
null_var=df.isnull().sum()/df.shape[0] *100
null_var
```

```python
drop_col=null_var[null_var>20].keys()
drop_col
```

```python
df2_drop_clm=df.drop(columns=drop_col)
```

```python
df3_num=df2_drop_clm.select_dtypes(include=['int64','float64'])
```

```python
df3_num.isnull().sum()
```

```python
missing_num_var=[var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_var
```

Data Distribution

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={'linewidth':5,'color':'r'})
```

When data is normally distributed use mean maximum times
In normal distribution we can use mean,median and mode

```python
df4_num_mean=df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={'linewidth':8,'color':'r'},label='Original')
    sns.distplot(df4_num_mean[var],bins=20,kde_kws={'linewidth':5,'color':'g'},label='Mean')
    plt.legend()
```

```python
df4_num_median=df3_num.fillna(df3_num.median())
df4_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,hist=False,kde_kws={'linewidth':8,'color':'r'},label='Original')
    sns.distplot(df4_num_mean[var],bins=20,hist=False,kde_kws={'linewidth':5,'color':'g'},label='Mean')
    sns.distplot(df4_num_median[var],bins=20,hist=False,kde_kws={'linewidth':3,'color':'y'},label='Median')
    plt.legend()
```

```python
for i,var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df3_num[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_num_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df4_num_mean[var])
```

```python
df_concat=pd.concat([df3_num[missing_num_var],df4_num_mean[missing_num_var],df4_num_mean[missing_num_var]]
                   ,axis=1)
```
