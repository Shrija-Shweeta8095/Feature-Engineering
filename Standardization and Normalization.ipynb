## Feature Scaling
### Standardization and Normalization

```python
import pandas as pd
import seaborn as sns
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df=sns.load_dataset("titanic")
```

```python
df
```

```python
df.head()
```

```python
df2=df[['survived','pclass','age','parch']]
```

```python
df2
```

```python
df2.head()
```

```python
df3=df2.fillna(df2.mean())
```

```python
X=df3.drop('survived',axis=1)
Y=df3['survived']
print("Shape of X:",X.shape)
print("Shape of Y:",Y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=51)
```

```python
print("Shape of X_train:",X_train.shape)
print("Shape of Y_train:",Y_train.shape)
print("Shape of X_test:",X_test.shape)
print("Shape of Y_test:",Y_test.shape)
```

```python
sc=StandardScaler()
sc.fit(X_train)
```

```python
sc.mean_
```

```python
sc.scale_  #Standard Deviation
```

```python
X_train.describe()
```

```python
X_train_sc=sc.transform(X_train)
X_test_sc=sc.transform(X_test)
```

```python
X_train_sc
```

```python
X_test_sc
```

```python
x_train_sc1=pd.DataFrame(X_train_sc,columns=["pclass","age","parch"])
x_test_sc1=pd.DataFrame(X_test_sc,columns=["pclass","age","parch"])
```

```python
x_train_sc1.head()
```

```python
x_test_sc1.head()
```

```python
x_train_sc1.describe().round(2)
```

```python
mmc=MinMaxScaler()
```

```python
mmc.fit(X_train)
```

```python
X_train_mmc=mmc.transform(X_train)
X_test_mmc=mmc.transform(X_test)
```

```python
X_train_mmc
```

```python
x_train_mmc1=pd.DataFrame(X_train_mmc,columns=["pclass","age","parch"])
x_test_mmc1=pd.DataFrame(X_test_mmc,columns=["pclass","age","parch"])
```

```python
x_train_mmc1.describe().round(2)
```
